## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here

# select all the info from the table named 'master'
QUERY = """
SELECT * 
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY, conn)

import matplotlib.pyplot as plt

# Plot the histogram
plt.hist(df["ConvertedComp"], bins=20)

# Add labels and title
plt.xlabel("ConvertedComp")
plt.ylabel("Frequency")
plt.title("Histogram of ConvertedComp")

# Show the plot
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

df.plot(kind='scatter', x='Age', y='WorkWeekHrs') # More parameters can be added in plot(), e.g., figsize=(10, 6), color='darkblue', etc.

# Add labels and title
plt.title('Scatter plot of Age and WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')

# Show the plot
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python

# A bubble plot is a variation of the scatter plot that displays three dimensions of data (x, y, s). 
# The data points are replaced with bubbles, and the size of the bubble is determined by the third variable s, also known as the weight. 
# In maplotlib, we can pass in an array or scalar to the parameter s to plot(), that contains the weight of each point.

# CodeRevHrs: On average, how many hours per week do you spend on code review?
# WorkWeekHrs: On average, how many hours per week do you work?
# Age: What is your age (in years)? If you prefer not to answer, you may leave this question blank.

# Solution 1:

df.plot(kind='scatter', x='WorkWeekHrs', y='CodeRevHrs', s=df['Age']*3) # More parameters can be added in plot(), e.g., figsize=(10, 6), color='darkblue', etc.

# Add labels and title
plt.title('Bubble plot of WorkWeekHrs and CodeRevHrs, use Age column as bubble size')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')

# Show the plot
plt.show()
```

```python
# Solution 2:

plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=df['Age']*3, alpha=0.5)

# Labels and title
plt.xlabel('Work Week Hours')
plt.ylabel('Code Review Hours')
plt.title('Bubble Plot of Work Week Hours vs Code Review Hours')
plt.grid(True)

# Show plot
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = """
SELECT * 
FROM DatabaseDesireNextYear
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY, conn)
df
```

```python
# Select "DatabaseDesireNextYear" and create "RespondentCount" from the table named 'DatabaseDesireNextYear'.

QUERY = """
SELECT DatabaseDesireNextYear, COUNT(Respondent) as RespondentCount
FROM DatabaseDesireNextYear
group by DatabaseDesireNextYear
order by RespondentCount desc
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY, conn)

df
```

```python

# Create pie chart:

df['RespondentCount'].head().plot(kind='pie',
                            figsize = (5, 6),
                            autopct = '%1.1f%%', # add in percentages, autopct create %
                            startangle = 90, # start angle 90° (PostgreSQL), startangle represents starting point
                            #shadow = True, # add shadow
                            labels = df['DatabaseDesireNextYear']
                            )

# Add title
plt.title('Pie chart of the top 5 databases that respondents wish to learn next year')
plt.axis('equal') # Sets the pie chart to look like a circle.

# Show the plot
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# select WorkWeekHrs, CodeRevHrs, and Age of age group 30 to 35 from the table named 'master'

QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE Age BETWEEN 30 AND 35
"""

# use WHERE Age>=30 and Age<=35 is also okay above

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY, conn)
df
```

```python
# Solution 1:
# this is to draw a stacked bar chart of WorkWeekHrs medians and CodeRevHrs medians for all respondents of age group 30 to 35

dfSol1 = df.groupby(by='Age').median()

# Plot stacked bar chart
dfSol1.plot(kind = 'bar', # 'bar' means stacked bar chart, 'area' means stacked line chart  
             stacked = True,
             figsize =(10, 5))  # pass a tuple (x, y) size

# Add title
plt.title('Stacked bar chart of median WorkWeekHrs and CodeRevHrs for the age group 30 to 35')

# Show the plot. If don't use the following line of code, the plot can also be drawn, try it!
plt.show()
```

```python
# Solution 2:
# this is to draw a stacked bar chart of WorkWeekHrs medians and CodeRevHrs medians for all respondents of age group 30 to 35

LstWorkWeekHrsMed = list()
LstCodeRevHrsMed = list()

age = 30
while age <= 35:
    WorkWeekHrsMed = df[df['Age'] == age]['WorkWeekHrs'].median() # the final plot below may look boring as all the WorkWeekHrs medians are the same, try to use mean() method here!  
    LstWorkWeekHrsMed.append(WorkWeekHrsMed)
    CodeRevHrsMed = df[df['Age'] == age]['CodeRevHrs'].median() # the final plot below may look boring as all the CodeRevHrs medians are the same, try to use mean() method here!
    LstCodeRevHrsMed.append(CodeRevHrsMed)
    age = age + 1

MedDict = {'WorkWeekHrs Median': WorkWeekHrsMed, 'CodeRevHrs Median': LstCodeRevHrsMed}
dfSol2 = pd.DataFrame(MedDict)

# Plot stacked bar chart
dfSol2.plot(kind = 'bar', # 'bar' means stacked bar chart, 'area' means stacked line chart  
             stacked = True,
             figsize =(10, 5))  # pass a tuple (x, y) size

# Add title
plt.title('Stacked bar chart of median WorkWeekHrs and CodeRevHrs for the age group 30 to 35')

# Show the plot. If don't use the following line of code, the plot can also be drawn, try it!
plt.show()

```

```python
# Practice drawing a stacked line chart of WorkWeekHrs and CodeRevHrs for all respondents of age group 18.

# select WorkWeekHrs and CodeRevHrs of age group 18 from the table named 'master'
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age = 18
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY, conn)
df
```

```python

# this is to draw a stacked line chart of WorkWeekHrs and CodeRevHrs for all respondents of age group 18

# Plot stacked line chart
df.plot(kind='area',            # 'barh' means horizontal stacked bar chart, 'bar' means vertical stacked bar chart, 'area' means stacked line chart. You can just remove the 'kind' parameter and a line chart (kind='line') will be drawn by default.
             stacked=True,
             figsize=(10, 5))  # pass a tuple (x, y) size

# Add title
plt.title('Stacked line chart of WorkWeekHrs and CodeRevHrs for the age group 18')

# Show the plot. If don't use the following line of code, the plot can also be drawn, try it!
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# select MainBranch and create MainBranchCount from the table named 'master'
QUERY = """
SELECT MainBranch, COUNT(MainBranch) as MainBranchCount
FROM master
GROUP BY MainBranch
"""

#the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY, conn)
df
```

```python

# plot horizontal bar chart using column MainBranch (Matplotlib Artist Layer):
ax = df.plot(kind='barh', figsize=(6, 3))

# add xlabel, title, and column names
ax.set_xlabel('Count of MainBranch')
ax.set_title('Horizontal bar chart using column MainBranch')
ax.set_yticklabels(['developer by profession', 'not developer, write code sometimes'])

# add numbers on the edge of the bars
for i, v in enumerate(df['MainBranchCount']):
    ax.text(v + 3, i - 0.1, str(v), color='blue')

# Show the plot
plt.show()

```
